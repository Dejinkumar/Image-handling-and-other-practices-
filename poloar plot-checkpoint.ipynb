```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
subjects=['DWT','DCT','RLE','Block trucation']
Compression=[54.76,89.97,43.72,47.63]
AHEEnhancement=[86.98,82.67,70.50,75.77]
MOEnhancement=[72.76,70,69.85,69.33]
```

```python
angles=np.linspace(0,2*np.pi,len(subjects), endpoint=False)
print(angles)
```

```python
angles=np.concatenate((angles,[angles[0]]))
subjects.append(subjects[0])
Compression.append(Compression[0])
AHEEnhancement.append(AHEEnhancement[0])
MOEnhancement.append(MOEnhancement[0])
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(projection='polar',frameon=False)
ax.plot(angles,Compression)
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)
#basic plot
ax.plot(angles,Compression, 'o--', color='g', label='Compression')
#fill plot
ax.fill(angles, Compression, alpha=0.25, color='g')
#Add labels
ax.set_thetagrids(angles * 180/np.pi, subjects)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(111, polar=True,frameon=False)

ax.plot(angles,Compression, 'o-',ms=15,mfc='white',mec='black', color='g', linewidth=5, label='Compression')
#ax.fill(angles, Compression, alpha=0.25, color='g')

ax.plot(angles,AHEEnhancement, 'o-',ms=15,mfc='white',mec='black', color='orange', linewidth=5, label='AHE Enhancement')
#ax.fill(angles, AHEEnhancement, alpha=0.25, color='orange')

ax.plot(angles,MOEnhancement, 'o-',ms=15,mfc='white',mec='black', color='red', linewidth=5, label='MOEnhancement')
#ax.fill(angles, MOEnhancement, alpha=0.25, color='orange')

ax.set_thetagrids(angles * 180/np.pi, subjects)
plt.grid(True,color='black',alpha=0.2)
plt.tight_layout()
plt.legend()
plt.show()
```
